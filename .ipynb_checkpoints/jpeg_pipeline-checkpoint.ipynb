# Pipeline JPEG

### Nous implémenterons le pipeline permettant de compresser une image en JPEG. Les étapes suivent celles décrites dans l'énoncé de laboratoire.

Référence: https://www.hdm-stuttgart.de/~maucher/Python/MMCodecs/html/jpegUpToQuant.html

```python
import sys
import numpy as np
from scipy.fftpack import dct, idct
import cv2
import matplotlib.pyplot as plt
from dahuffman import HuffmanCodec
```

```python
img = plt.imread('carlos_out.png')
# Initial size of the image
CHAN_NBR = len(img[0][0]) - 1 # Without the alpha channel
IMG_WIDTH = len(img[0])
IMG_HEIGHT = len(img)
ORIGINAL_SIZE = CHAN_NBR * IMG_WIDTH * IMG_HEIGHT
print("Initial size of the image [bytes]:", ORIGINAL_SIZE)
plt.imshow(img[:,:])
plt.show()
```

# Encodage de l'image

### Étape 1: Transformation et sous-échantillonage des canaux RGB en canaux YCbCr

```python


# On convertit dans le domaine YCbCr
img_YBR = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
img_Cr = img_YBR[:,:,1]
img_Cb = img_YBR[:,:,2]

print(len(img_YBR))

# On ignore les valeurs de Cr et de Cb à chaque colonne et rangée impaires (ou paires)
# a[start:end:step] -> Quand on ne spécifie pas start et end, on supprime chaque multiple de step
img_Cr_subsampled = img_Cr[::2,::2]
img_Cb_subsampled = img_Cb[::2,::2]

img_YBR = np.array((img_YBR[:,:,0], img_Cr_subsampled, img_Cb_subsampled))
for i, channel in enumerate(['Luminance', 'Canal rouge', 'Canal bleu']):
    plt.title(channel)
    plt.imshow(img_YBR[:][:][i])
    plt.show()
```

### Étape 2: Le découpage en blocs de 8x8 pixels

```python
BLOCK_SIZE = 8
blocks = []
IMG_SIZE = [(len(elem),len(elem[0])) for elem in img_YBR]

for i in range(len(img_YBR)):
    blocks.append([])
    if(len(img_YBR[i]) % BLOCK_SIZE != 0):
        raise Exception("Les dimensions de l'image ne sont pas des multiples de la taille du bloc.")
    for j in range((int)(len(img_YBR[i])/BLOCK_SIZE)):
        if(len(img_YBR[i][0]) % BLOCK_SIZE != 0):
            raise Exception("Les dimensions de l'image ne sont pas des multiples de la taille du bloc.")
        for k in range((int)(len(img_YBR[i][0])/BLOCK_SIZE)):
            block = img_YBR[i][j*BLOCK_SIZE:j*BLOCK_SIZE + BLOCK_SIZE, k*BLOCK_SIZE: k*BLOCK_SIZE + BLOCK_SIZE]
            blocks[i].append(block)
```

### Étape 3 : Transformée en cosinus discrète (DCT) 

```python
def transform_to_freq_dom(block):
    return dct(dct(block, axis=0, norm='ortho'), axis=1, norm='ortho')
```

### Étape 4: Quantification

```python
quantization_matrix = np.matrix(
    '16 11 10 16 24 40 51 61;   \
    12 12 14 19 26 58 60 55;    \
    14 13 16 24 40 57 69 56;    \
    14 17 22 29 51 87 80 62;    \
    18 22 37 56 68 109 103 77;  \
    24 35 55 64 81 104 103 92;  \
    49 64 78 77 103 121 120 101;\
    72 92 95 98 112 100 103 99'
)

def quantize_block(dct_block):
    global quantization_matrix
    return np.round(np.divide(dct_block, quantization_matrix))
```

### Étape 5: Zigzag

```python
vectorization_matrix = np.matrix(
   '0   1  5  6 14 15 27 28;\
    2   4  7 13 16 26 29 42;\
    3   8 12 17 25 30 41 43;\
    9  11 18 24 31 40 44 53;\
    10 19 23 32 39 45 52 54;\
    20 22 33 38 46 51 55 60;\
    21 34 37 47 50 56 59 61;\
    35 36 48 49 57 58 62 63'
)


def convert_to_1D(quantized_block):
    global vectorization_matrix
    vectorized_block = []
    for i in range(BLOCK_SIZE**2):
        position = np.where(vectorization_matrix == i)
        position = (position[0][0], position[1][0])
        vectorized_block.append(quantized_block[position[0], position[1]])
    return vectorized_block
```

### Étape 6: Encodage de Huffman et RLE
On utilise pas de codage prédictif pour le DCT par mesure de simplicité. On applique donc aussi Huffman et RLE pour ce codage.

```python
def huffman_rle(vectorized_block):
    codec_block = HuffmanCodec.from_data(list(filter(lambda x: x != 0, vectorized_block)))
    zeros_cnt = 0
    encoded_huffman_rle = []
    for symb in vectorized_block:
        if symb != 0:
            encoded_symb = codec_block.encode([symb])
            encoded_huffman_rle.append((zeros_cnt, encoded_symb))
            zeros_cnt = 0
        else:
            zeros_cnt += 1
    return (encoded_huffman_rle, codec_block)
```

### Exécution des étapes 3 à 6 pour chaque bloc

```python
compressed_img = []

for i in range(len(blocks)):
    compressed_img.append([])
    for block in blocks[i]:
        block = (block * 255) - 128
        dct_block = transform_to_freq_dom(block)
        quantized_block = quantize_block(dct_block)
        vectorized_block = convert_to_1D(quantized_block)
        huffman_rle_result = huffman_rle(vectorized_block)
        compressed_img[i].append(huffman_rle_result)
```

### Étape 7: Calcul du taux de compression:

```python
BYTES_PER_ENCODING = 2

compressed_img_size = 0
for compressed_channel in compressed_img:
    for compressed_block in compressed_channel:
        compressed_img_size += len(compressed_block[0]) * BYTES_PER_ENCODING
        
compression_rate = 1 - (compressed_img_size/ORIGINAL_SIZE)
print('Compression rate [bytes]: ', compression_rate)
```

# Décodage de l'image: 

### Étape 6: Encodage de Huffman et RLE

```python
def unhuffman_rle(encoded_huffman_rle):
    vectorized_block = np.zeros(BLOCK_SIZE**2)
    zeros_cnt = 0
    for encoded_symb in encoded_huffman_rle[0]:
        zeros_cnt += encoded_symb[0]
        decoded_symb = encoded_huffman_rle[1].decode(encoded_symb[1])
        vectorized_block[zeros_cnt] = decoded_symb[0]
        zeros_cnt += 1
    return vectorized_block
```

### Étape 5: Zigzag

```python
def convert_to_2D(vectorized_block):
    quantized_block = np.zeros([BLOCK_SIZE, BLOCK_SIZE])
    for i in range(len(vectorized_block)):
        x = i//BLOCK_SIZE
        y = i%BLOCK_SIZE
        index = vectorization_matrix[x, y]
        quantized_block[x][y] = vectorized_block[index]
    return quantized_block
```

### Étape 4: Quantification

```python
def dequantize_block(quantized_block):
    global quantization_matrix
    return np.round(np.multiply(quantized_block, quantization_matrix))
```

### Étape 3: Transformée en cosinus discrète

```python
def transform_to_time_dom(dct_block):
    return idct(idct(dct_block, axis=0, norm='ortho'), axis=1, norm='ortho')
```

### Exécution des étapes 6 à 3 pour la décompression

```python
blocks = []

for i, compressed_channel in enumerate(compressed_img):
    blocks.append([])
    for compressed_block in compressed_channel:
        vectorized_block = unhuffman_rle(compressed_block)
        quantized_block = convert_to_2D(vectorized_block)
        dct_block = dequantize_block(quantized_block)
        block = transform_to_time_dom(dct_block) +128
        blocks[i].append(block)
```

### Étape 2: Division par blocs

```python
img_YBR = []

print(IMG_SIZE[0], IMG_SIZE[1], IMG_SIZE[2])

for i, block_channel in enumerate(blocks):
    img_YBR.append(np.zeros([IMG_SIZE[i][0],IMG_SIZE[i][1]]))
    # On regarde ici pour chaque channel
    NB_BLOCKS_PER_ROW = int(IMG_SIZE[i][1] / BLOCK_SIZE)
    NB_BLOCKS_PER_COLUMN = int(IMG_SIZE[i][0] / BLOCK_SIZE)
    
    column_offset = 0;
    for j in range(NB_BLOCKS_PER_COLUMN):
        for k in range(NB_BLOCKS_PER_ROW):
            img_YBR[i][j*BLOCK_SIZE: j*BLOCK_SIZE + BLOCK_SIZE, k*BLOCK_SIZE:k*BLOCK_SIZE+BLOCK_SIZE] = (block_channel[k + j * NB_BLOCKS_PER_ROW])
        
        
print(len(img_YBR[1][0]) )   
```

### Étape 1: Sous échantillonage de la chrominance

```python
# Get initial size of image by looking at the luminance channel
N = len(img_YBR[:][:][0])
img_out = np.empty([3, N, N])
img_out[0] = img_YBR[:][:][0]

img_out_inversed = np.empty((N,N,3), dtype=np.uint8)
img_out_inversed[:,:,0] = [[img_YBR[0][y][x] for x in range(0,N)] for y in range(0,N)]

for idx, img_chrominance_subsampled in enumerate([img_YBR[:][:][1], img_YBR[:][:][2]]):
    img_chrominance_subsampled = cv2.resize(img_chrominance_subsampled, (N, N))
    img_out_inversed[:,:, idx+1] = [[(img_chrominance_subsampled[y][x]) for x in range(0,N)] for y in range(0,N)]
    
img_done = cv2.cvtColor(img_out_inversed, cv2.COLOR_YCrCb2RGB)

plt.imsave("carlos_out.png", img_done)

plt.imshow(img_done)
plt.show()
```
